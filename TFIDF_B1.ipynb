```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
doc1 =["Prakash is a good boy","Ramesh is good boy","Vinay is good boy"]
vect =CountVectorizer()
vectcount =vect.fit_transform(doc1)
print(vect.get_feature_names())
vectcount.toarray()
vectcount.toarray().sum(axis=0)
```

```
tfid = TfidfVectorizer(use_idf=True)
tfidvect = tfid.fit_transform(doc1)
print(tfid.get_feature_names())
tfidvect.toarray()
```
